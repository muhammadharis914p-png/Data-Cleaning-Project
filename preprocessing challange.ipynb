```python
from sklearn.impute import SimpleImputer

def impute_missing_values(df):
    """
    Impute missing values in a DataFrame based on the data type of each column using scikit-learn's SimpleImputer.

    Parameters:
    - df: pandas DataFrame

    Returns:
    - df_imputed: DataFrame with missing values imputed

    - Strategy str, default=’mean’
    The imputation strategy.

    If “mean”, then replace missing values using the mean along each column. Can only be used with numeric data.

    If “median”, then replace missing values using the median along each column. Can only be used with numeric data.

    If “most_frequent”, then replace missing using the most frequent value along each column.
    Can be used with strings or numeric data. If there is more than one such value, only the smallest is returned.

    If “constant”, then replace missing values with fill_value. Can be used with strings or numeric data.
    """

    # Separate columns into categorical and numerical
    categorical_columns = df.select_dtypes(include='object').columns
    numerical_columns = df.select_dtypes(exclude='object').columns

    # Impute missing values for categorical columns with the mode
    categorical_imputer = SimpleImputer(strategy='most_frequent')
    df_imputed = df.copy()  # Create a copy to avoid modifying the original DataFrame
    df_imputed[categorical_columns] = categorical_imputer.fit_transform(df_imputed[categorical_columns])

    # Impute missing values for numerical columns with the mean
    numerical_imputer = SimpleImputer(strategy='mean')
    df_imputed[numerical_columns] = numerical_imputer.fit_transform(df_imputed[numerical_columns])

    return df_imputed
```

```python
# Impute missing values using the function on Original dataset df
df_imputed_data = impute_missing_values(df)
```

```python
from sklearn.preprocessing import LabelEncoder
def encode_columns(df_cleaned_columns, label_encode_cols=None, onehot_encode_cols=None):
    """
    Perform label encoding and one-hot encoding on specified columns of a DataFrame.

    Parameters:
    - df: pandas DataFrame
    - label_encode_cols: List of column names to be label encoded
    - onehot_encode_cols: List of column names to be one-hot encoded

    Returns:
    - encoded_df: DataFrame with specified columns encoded
    """

    # Copy the original DataFrame to avoid modifying it
    encoded_df = df_cleaned_columns.copy()

    # Label encoding
    if label_encode_cols:
        label_encoder = LabelEncoder()
        for col in label_encode_cols:
            if col in encoded_df.columns:
                encoded_df[col] = label_encoder.fit_transform(encoded_df[col])

    # One-hot encoding
    if onehot_encode_cols:
        encoded_df = pd.get_dummies(encoded_df, columns=onehot_encode_cols, dtype=int)

    return encoded_df 
```

```python
# Example columns are choosen for the Demo purpose

# Specify columns for label encoding and one-hot encoding
label_encode_columns = ['country','director_name']
onehot_encode_columns = ['content_rating']

# Apply the encoding function
df_encoded = encode_columns(df_cleaned_rows, label_encode_cols=label_encode_columns, onehot_encode_cols=onehot_encode_columns)

# Set the option to display all columns
pd.set_option('display.max_columns', None)

# Printing the dataset
display(df_encoded)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
def scale_columns(df_encoded, scale_cols=None, scaler=None):
    """
    Perform scaling on specified columns of a DataFrame using a specified scaler from scikit-learn.

    Parameters:
    - df: pandas DataFrame
    - scale_cols: List of column names to be scaled
    - scaler: Scaler object from scikit-learn (e.g., StandardScaler, MinMaxScaler, RobustScaler)

    Returns:
    - scaled_df: DataFrame with specified columns scaled
    """

    # Copy the original DataFrame to avoid modifying it
    scaled_df = df_encoded.copy()

    # Scaling
    if scale_cols and scaler:
        scaler_instance = scaler()
        scaled_df[scale_cols] = scaler_instance.fit_transform(scaled_df[scale_cols])

    return scaled_df
```

```python
# Example columns are choosen for  DEMO purpose
# Specify numerical columns columns for scaling 
scale_columns_list = ['director_name']

# Specify the scaler you want to use (e.g., StandardScaler, MinMaxScaler, RobustScaler)
chosen_scaler = StandardScaler

# Apply the scaling function
df_scaled = scale_columns(df_encoded, scale_cols=scale_columns_list, scaler=chosen_scaler)

# Display the scaled DataFrame
pd.set_option('display.max_columns', None)

# Printing the dataset
display(df_scaled)
```
